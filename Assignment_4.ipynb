```python
# Task 3: Handling Null Values
# Create a new DataFrame with some null values and fill them

df_null = pd.DataFrame({
    "Name": ["Alice", "Bob", None, "David"],
    "Age": [25, None, 35, 28],
    "City": ["New York", "San Francisco", "Los Angeles", None]})

# Fill null values
# Fill missing names with "Unknown"
df_null["Name"] = df_null["Name"].fillna("Unknown")

# Fill missing ages with the average age
df_null["Age"] = df_null["Age"].fillna(df_null["Age"].mean())

# Fill missing cities with "Unknown City"
df_null["City"] = df_null["City"].fillna("Unknown City")

print("\nDataFrame after filling null values:")
print(df_null)

```

```python
# Task 5: Concatenation and Merging
# Concatenate two DataFrames vertically and merge them horizontally

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
df3 = pd.DataFrame({'C': [9, 10], 'D': [11, 12]})

# Vertical concatenation (stack rows)
df_vertical = pd.concat([df1, df2], axis=0)
print("Vertical Concatenation:")
print(df_vertical)

# Horizontal merge (side-by-side)
df_horizontal = pd.concat([df1, df3], axis=1)
print("\nHorizontal Merge:")
print(df_horizontal)
```

```python
# Task 8: Functions and Lambda
# Create a function to calculate the square of a number.

def square(num):
    return num * num

print("Square using function:", square(5))

# Lambda function to calculate the square
square_lambda = lambda num: num * num

print("Square using lambda:", square_lambda(5))
```

```python
# Iterator for the first 5 even numbers

class EvenIterator:
    def __init__(self):
        self.num = 0   # start at 0
        self.count = 0 # how many evens we've produced

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < 5:
            even = self.num
            self.num += 2
            self.count += 1
            return even
        else:
            raise StopIteration

# Using the iterator
even_iter = EvenIterator()
print("Iterator output:")
for n in even_iter:
    print(n)
```

```python
# Task 10: Map, Reduce, and Filter
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# 1. Use map to square all numbers
squared = list(map(lambda x: x * x, numbers))
print("Squared numbers:", squared)

# 2. Use reduce to find the product of all numbers
product = reduce(lambda x, y: x * y, numbers)
print("Product of numbers:", product)

# 3. Use filter to get only even numbers
evens = list(filter(lambda x: x % 2 == 0, numbers))
print("Even numbers:", evens)
```

```python
# Task 11: Object-Oriented Programming - Creating a Class
# Create a class called 'Rectangle' with attributes 'length' and 'width'.
# Include methods to calculate area and perimeter.

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

# Testing the class
rect = Rectangle(5, 3)
print("Area:", rect.area())
print("Perimeter:", rect.perimeter())
```

```python
# Task 12: Pandas Data Analysis
df_employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]
})

# 1. Find the average salary by department
avg_salary = df_employees.groupby("Department")["Salary"].mean()
print("Average salary by department:")
print(avg_salary)

# 2. Get the names of employees with salary > 60000
high_salary = df_employees[df_employees["Salary"] > 60000]["Name"]
print("\nEmployees with salary > 60000:")
print(list(high_salary))

# 3. Add a new column 'Bonus' which is 10% of salary
df_employees["Bonus"] = df_employees["Salary"] * 0.10
print("\nDataFrame with Bonus column:")
print(df_employees)
```
